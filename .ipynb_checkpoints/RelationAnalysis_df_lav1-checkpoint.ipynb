### Import df_lav1

```python
file_path = r'data_output\df_lav1.csv'
df_lav1 = pd.read_csv(file_path)
df_lav1["New_Risk"].value_counts()
```

```python
categorical_columns_to_force = ['Altr_CUP_COD_NATURA', 'Altr_CUP_COD_SETTORE', 'New_Risk']
df_lav1[categorical_columns_to_force] = df_lav1[categorical_columns_to_force].astype('object')
```

```python
# Selezionare solo le colonne numeriche
numeric_cols = df_lav1.select_dtypes(include=['int64', 'float64'])

# Calcolare la matrice di correlazione
correlation_matrix = numeric_cols.corr()

# Stampare la matrice di correlazione
print("Matrice di correlazione:")
print(correlation_matrix)

# Visualizzare la heatmap delle correlazioni
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap delle correlazioni")
plt.show()

# Scatterplot per una variabile numerica rispetto al target
sns.pairplot(df_lav1, vars=numeric_cols.columns, hue='New_Risk', diag_kind='kde')
plt.show()
```

```python
# Seleziona solo le colonne categoriali
categorical_cols = df_lav1.select_dtypes(include=['object']).columns

# Analisi delle variabili categoriali rispetto al target
target_col = 'New_Risk'  # Specifica la colonna del target
categorical_cols_with_target = list(categorical_cols) + [target_col]

# Tabella di contingenza per ogni variabile categoriale rispetto al target
for col in categorical_cols:
    print(f"\nAnalisi della variabile categoriale: {col}")
    
    # Creare la tabella di contingenza
    contingency_table = pd.crosstab(df_lav1[col], df_lav1[target_col])
    print("Tabella di contingenza:")
    print(contingency_table)
    
    # Test chi-quadrato
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print(f"\nTest chi-quadrato:\nChi2 = {chi2:.2f}, p-value = {p:.4f}, Degrees of Freedom = {dof}")
    
    if p < 0.05:
        print("La relazione tra questa variabile e il target è statisticamente significativa.")
    else:
        print("Non è stata rilevata una relazione statisticamente significativa.")
    
    # Visualizzare la relazione con un grafico
    contingency_table.plot(kind='bar', stacked=True, figsize=(8, 5))
    plt.title(f"Distribuzione del target rispetto a {col}")
    plt.xlabel(col)
    plt.ylabel("Frequenza")
    plt.legend(title=target_col)
    plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import itertools
import seaborn as sns
import matplotlib.pyplot as plt

# Selezionare solo le colonne categoriali
categorical_cols = df_lav1.select_dtypes(include=['object']).columns

# Inizializzare un dizionario per salvare i risultati
chi2_results = []

# Test chi-quadrato tra tutte le coppie di variabili categoriali
for col1, col2 in itertools.combinations(categorical_cols, 2):  # Genera tutte le coppie di colonne
    # Creare la tabella di contingenza
    contingency_table = pd.crosstab(df_lav1[col1], df_lav1[col2])
    
    # Eseguire il test chi-quadrato
    try:
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        # Salvare i risultati
        chi2_results.append({
            'Variable 1': col1,
            'Variable 2': col2,
            'Chi2': chi2,
            'p-value': p
        })
    except ValueError:
        # Se ci sono problemi con la tabella di contingenza, passare oltre
        pass

# Creare un DataFrame con i risultati
chi2_df = pd.DataFrame(chi2_results)

# Filtrare solo le correlazioni significative (p-value < 0.05)
significant_correlations = chi2_df[chi2_df['p-value'] < 0.05]

# Visualizzare i risultati come tabella
print(significant_correlations)

# Visualizzare una heatmap delle variabili significative
if not significant_correlations.empty:
    # Creare una matrice per la heatmap
    heatmap_data = significant_correlations.pivot(index="Variable 1", columns="Variable 2", values="Chi2")
    
    # Creare la heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(heatmap_data, annot=True, cmap="coolwarm", cbar=True, fmt=".2f")
    plt.title("Heatmap delle Correlazioni Significative tra Variabili Categoriali")
    plt.show()
else:
    print("Non ci sono correlazioni significative tra le variabili categoriali.")
```
